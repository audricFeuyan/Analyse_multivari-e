## Définition des fonctions d'analyse

```python
def display_data_info(data):
    data.info()
    data.dtypes
    
def display_null_data_colum_ratio(data):
    print(((data.isna().sum()/data.shape[0])*100).sort_values(ascending=False))
    
def delete_duplicated_data(data, tokeep="first", columns=None):
    return data.drop_duplicates(subset=columns, keep=tokeep)

def outliers_proportion(data):
    nbdata = data.shape[0]
    Min=data.min()
    Max=data.max()
    Q1=data.quantile(0.25)
    Q3=data.quantile(0.75)
    IQR=Q3-Q1
    outliers_low=(Q1-1.5*IQR)
    outliers_high=(Q3+1.5*IQR)
    print(((data<outliers_low) | (data>outliers_high)).sum()/nbdata*100)
    
```

### Analyse multivariée

```python
def plot_agregate_boxplot(data, x_column, y_column, x_label, y_label, plot_title):
    X = x_column
    Y = y_column


    modalites = data[X].unique()
    modalites.sort()
    groupes = []
    for m in modalites:
        groupes.append(data[data[X]==m][Y])


    medianprops = {'color':'black'}
    meanprops = {'marker':'o', 'markeredgecolor':'black','markerfacecolor':'firebrick'}

    plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
                vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plot_title)
    plt.show()
    

def anova_coef(X_values_qualitative, Y_value_quantitative):    
    moyenne_y = Y_value_quantitative.mean()
    classes = []
    for classe in X_values_qualitative.unique():
        yi_classe = Y_value_quantitative[X_values_qualitative==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in Y_value_quantitative])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
```

```python
plot_agregate_boxplot(foods_france_final, "additives_n", "nutrition-score-fr_100g", "Score nutritionnel", "Additifs", "Score nutritionnel/Additifs")
```

#### PCA

```python
foods_france_brut = foods_france_final.select_dtypes(exclude='object')
```

```python
from sklearn.preprocessing import StandardScaler

# On centre et réduit les données
foods_france_norm = StandardScaler().fit_transform(foods_france_brut)
foods_france_norm = pd.DataFrame(foods_france_norm, columns=foods_france_brut.columns)
foods_france_norm.sample(5)
```

```python
from sklearn.decomposition import PCA

# On apprends des données pour calculer les coefficients de chaque fonction
pca = PCA().fit(foods_france_norm)
pca.components_
```

```python
pca_components = pd.DataFrame(pca.components_,
                      index=['PCA'+str(i+1) for i in range(foods_france_norm.shape[1])],
                      columns=foods_france_norm.columns).T
pca_components
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
heatmap=sns.heatmap(pca_components,annot=True, fmt=".3f", cmap='coolwarm')
heatmap.set_title('Coefficient des composantes principales', fontdict={'fontsize':18}, pad=16);
```

```python
#Important de chaque composante
print(pca.explained_variance_ratio_*100)
```

```python
# Projection sur les composantes principales
PCA().fit(foods_france_norm)
foods_france_pca = pca.transform(foods_france_norm)
foods_france_pca
```

```python
# Projection des données
foods_france_projected=pd.DataFrame(foods_france_pca,columns=['PCA'+str(i+1) for i in range(foods_france_brut.shape[1])])
foods_france_projected.head(5)
```

```python
# Corrélation après projection
heatmap=sns.heatmap(foods_france_projected.corr(),annot=True, fmt=".2f", cmap='coolwarm')
heatmap.set_title('Principal components coefficients', fontdict={'fontsize':18}, pad=16); 
```

```python
# Les composantes avec une inertie de pkaiser garantissent une bonne précision
pkaiser=100/foods_france_brut.shape[1]
pkaiser
```

```python
infopca=pd.DataFrame()
infopca["variables"]=['PCA'+str(i+1) for i in range(foods_france_norm.shape[1])]
infopca["importance"]=pca.explained_variance_ratio_*100
infopca.set_index("variables", inplace=True)
plt.rcParams['figure.figsize'] = [8, 8]
infopca.plot(kind="bar")
plt.show()
```

A ce stade, en considérant le critère de Kaiser, les composantes PCA1, PCA2, à PCA11 sont nécessaires pour expliquer nos variables.

```python
#Variances expliquées
important_degres = pca.explained_variance_ratio_*100

#Projection de l'éboulis des valeurs propres
plt.figure(figsize=(12,8))
plt.bar(np.arange(len(important_degres))+1, important_degres)
plt.plot(np.arange(len(important_degres))+1, important_degres.cumsum(),c="red",marker='o')
plt.xlabel("rang de l'axe d'inertie")
plt.ylabel("pourcentage d'inertie")
plt.title("Eboulis des valeurs propres")
plt.show(block=False)
```

L'éboulis des valeurs propres nous permet de dire que toutes les composantes sont nécessaires pour expliquer nos variables.
